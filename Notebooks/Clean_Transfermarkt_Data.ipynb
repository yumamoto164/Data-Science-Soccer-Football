```python
import pandas as pd
import numpy as np
```

```python
premier_league_df = pd.read_csv('premier_league_market_value.csv', index_col=0)
la_liga_df = pd.read_csv('la_liga_market_value.csv', index_col=0)
serie_a_df = pd.read_csv('serie_a_market_value.csv', index_col=0)
bundesliga_df = pd.read_csv('bundesliga_market_value.csv', index_col=0)
ligue_1_df = pd.read_csv('ligue_1_market_value.csv', index_col=0)
```

```python
ligue_1_df
```

```python
# Combine all 5 datasets together
combined_df = pd.concat([premier_league_df, la_liga_df, serie_a_df, bundesliga_df, ligue_1_df])
combined_df
```

```python
# Delete all rows with market values of 0
missing_values_index = combined_df[combined_df['Values'] == '0'].index
combined_df.drop(missing_values_index, inplace=True)
combined_df
```

```python
# Remove $ sign and '\xa0\xa0' from values column
for index, row in combined_df.iterrows():
    row['Values'] = row['Values'][1:-2]

combined_df
```

```python
combined_df.Values.replace(r'[Th.m]+$', '', regex=True).astype(float)
```

```python
combined_df.Values.str.extract(r'[\d\.]+([Th.m]+)', expand=False).unique()
```

```python
combined_df.Values.str.extract(r'[\d\.]+([Th.m]+)', expand=False).replace(['Th.', 'm'], [10**3, 10**6]).astype(int)
```

```python
combined_df.Values = (combined_df.Values.replace(r'[Th.m]+$', '', regex=True).astype(float) * \
              combined_df.Values.str.extract(r'[\d\.]+([Th.m]+)', expand=False).replace(['Th.', 'm'], [10**3, 10**6]).astype(int))
```
