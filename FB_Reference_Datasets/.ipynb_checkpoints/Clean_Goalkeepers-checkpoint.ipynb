```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('goalkeeping.csv')
df
```

```python
# Remove blank first row
df.drop(index=0, axis=0, inplace=True)
df = df.reset_index(drop=True)

# Clean Player Name Column
for index, row in df.iterrows():
    end = row['Player'].find('\\')
    df.loc[index, 'Player'] = row['Player'][0:end]
    
df
```

```python
df.describe()
```

```python
# Delete Penalty Stats
df = df.drop(['PKatt', 'PKA', 'PKsv', 'PKm', 'Save%.1'], axis=1)
df
```

```python
# Filla NaN values with 0
df = df.fillna(0)ions have new methods 'DataFrame.isna()' and 'DataFrame.notna()'

In [71]: df
Out[71]:
     a    b  c
0  NaN  7.0  0
df.describe()
```

```python
advanced_df = pd.read_csv('advanced_goalkeeping.csv')

# Remove blank first row and reset index
advanced_df.drop(index=0, axis=0, inplace=True)
advanced_df = advanced_df.reset_index(drop=True)
advanced_df = advanced_df.drop(['Att.1', 'Thr', 'Launch%', 'AvgLen', 'Att.2', 'Launch%.1', 'AvgLen.1',], axis=1)

# Clean Player Name Column
for index, row in advanced_df.iterrows():
    end = row['Player'].find('\\')
    advanced_df.loc[index, 'Player'] = row['Player'][0:end]
    
advanced_df.describe()
```

```python
combined_df = df.merge(advanced_df)
```

```python
combined_df
```

```python
# Load transfermarkt dataset
transfer_df = pd.read_csv("top5_league_market_values.csv", index_col=0)
transfer_df = transfer_df.rename(columns={"Players": "Player"})

# Delete Duplicate Players
transfer_df.drop_duplicates(subset ="Player", inplace = True)
print(transfer_df.nunique())
print(transfer_df.shape)
```

```python
duplicate_series = combined_df.duplicated(subset=['Player'].nunique()"Player", keep=False)
combined_df[duplicate_series]
```

```python
# From above delete Alfred Gomis (Dijon), Edouard Mendy (Rennes), Mathew Ryan (Arsenal)
combined_df.drop(index=[73, 114, 171], inplace=True)
```

```python
combined_df
```

```python
# Combine transfermarkt data and fb reference data
final_df = pd.merge(combined_df, transfer_df, on="Player")
final_df.fillna(0, inplace=True)
```
